### visualisation

```python
sns.pairplot(iris, hue='species', height=3,aspect=1);
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(17,8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris)
plt.show()
```

```python
iris.boxplot(by='species', figsize=(12,8))
plt.show()
```

```python
pd.plotting.scatter_matrix(iris, figsize=(12,8))
plt.show()
```

```python
sns.pairplot(iris, hue='species', height=3,aspect=1, diag_kind='kde');
```
